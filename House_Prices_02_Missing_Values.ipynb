```python
train_data.drop(columns='Id', inplace=True)
train_data.head()
```

### Checking for missing values in the data

```python
train_data.info()
```

#### Let's create a table that contains both the null values count and percentage for each variable of the dataset

```python
data_columns = [x for x in train_data.columns]
data_columns = pd.Series(data_columns, name='VariableName')

null_values_count = [len(train_data[train_data[x].isnull()==True]) for x in data_columns]
null_values_count = pd.Series(null_values_count, name='NullValues')

null_values_percent = [round((x / len(train_data))*100, 2) for x in null_values_count]
null_values_percent = pd.Series(null_values_percent, name='PercentNull')

null_values = pd.concat([data_columns, null_values_count, null_values_percent], axis=1)

null_values_table = null_values.loc[null_values['NullValues'] > 0].sort_values(by='NullValues', ascending=False)
null_values_table.reset_index(drop=True, inplace=True)
null_values_table
```

```python
null_values_table.to_csv('C:/Users/Satellite/Documents/1.DATA SCIENCE/KAGGLE/House Prices/null_values_table.csv')
```
