## Checking for missing values in the data

```python
train_data.info()
```

```python
train_data.columns
```

```python
data_columns = [x for x in train_data.columns]
data_columns = pd.Series(data_columns)
data_columns
```

```python
null_values_count = [len(train_data[train_data[x].isnull()==True]) for x in data_columns]
null_values_count = pd.Series(null_values_count)
null_values_count
```

```python
null_values = pd.concat([data_columns, null_values_count], axis=1)
null_values
```

```python
null_values_table = null_values.loc[null_values[1] > 0].sort_values(by=1, ascending=False)
null_values_table
```
