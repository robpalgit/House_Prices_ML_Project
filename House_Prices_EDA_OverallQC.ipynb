```python
import pandas as pd
from scipy.stats import pearsonr, spearmanr
```

```python
import matplotlib.pyplot as plt
import seaborn as sb
sb.set(style="whitegrid")
```

```python
from matplotlib import rcParams
rcParams['figure.figsize'] = 10, 8
```

```python
train_data = pd.read_csv('C:/Users/Satellite/Documents/1.DATA SCIENCE/KAGGLE/House Prices/train.csv')
train_data.drop(columns='Id', inplace=True)
train_data.head()
```

## OverallQual

```python
train_data['OverallQual'].value_counts()
```

```python
sb.countplot(train_data['OverallQual'])
```

```python
sb.regplot(x='OverallQual', y='SalePrice', data=train_data, scatter=True)
```

```python
spearmanr(train_data['OverallQual'], train_data['SalePrice'])
```

```python
pearsonr(train_data['OverallQual'], train_data['SalePrice'])
```

```python
sb.boxplot(x='OverallQual', y='SalePrice', data=train_data, palette='hls')
```
