```python
import pandas as pd
```

```python
train_data = pd.read_csv('C:/Users/Satellite/Documents/1.DATA SCIENCE/KAGGLE/House Prices/train.csv')
train_data.drop(columns='Id', inplace=True)
train_data.head()
```

```python
spearman_values = train_data.corr(method='spearman').loc['SalePrice'].sort_values(ascending = False).rename(index='SpearmanCorr')
data_columns = [x for x in spearman_values.index]

null_values_count = [len(train_data[train_data[x].isnull()==True]) for x in data_columns]
null_values_count = pd.Series(null_values_count, name='NullValues')

null_values_percent = [round((x / len(train_data))*100, 2) for x in null_values_count]
null_values_percent = pd.Series(null_values_percent, name='PercentNull')

spearman_corr = spearman_values.reset_index()

table_1 = pd.concat([spearman_corr, null_values_count, null_values_percent], axis=1).rename(columns={"index": "VariableName"})
table_1
```
